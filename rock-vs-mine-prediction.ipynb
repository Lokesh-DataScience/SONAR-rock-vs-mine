# Importing libraries

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

# Data collection and preprocessing

```python
#loading the datset to a pandas dataframe
sonar_data = pd.read_csv("Dataset/sonar data.csv",header=None)
```

```python
sonar_data.head()
```

```python
#number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()       #describe the statical methods for data
```

```python
sonar_data[60].value_counts()
```

M ---> Mine<br>
R ---> Rock

```python
sonar_data.groupby(60).mean()
```

```python
#separating data and labels
X = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

# Training and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,stratify= y ,random_state=1)
# When you use stratify in train_test_split, it’s like saying, 
# "Hey! Make sure when you split the data , each groups have the same types and amounts of data classes."
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

Model Training --> Logistic Regression

```python
lr_model = LogisticRegression(solver='liblinear')  # Use 'liblinear' for small datasets
```

```python
#Training the logistic regression model with training data
lr_model.fit(X_train,y_train)
```

# Model Evaluation

```python
#accuracy on training data
X_train_pred = lr_model.predict(X_train)
training_data_acc = accuracy_score(X_train_pred,y_train)
print("Accuracy on training data : ",training_data_acc*100)
```

```python
#accuracy on test data
X_test_pred = lr_model.predict(X_test)
test_data_acc = accuracy_score(X_test_pred,y_test)
print("Accuracy on test data : ",test_data_acc*100)
```

# Hyperparameter Tuning

```python
# Define hyperparameter grid
param_grid = {
    'C': np.logspace(-4, 4, 20),  # Regularization strength
    'penalty': ['l1', 'l2'],       # Type of regularization
}
```

```python
# Set up grid search
grid_search = GridSearchCV(estimator=lr_model, param_grid=param_grid,refit=True ,cv=5, scoring='accuracy', verbose=1)
```

```python
# Fit grid search to the training data
grid_search.fit(X_train, y_train)
```

```python
# Best hyperparameters
print("Best Hyperparameters: ", grid_search.best_params_)

# Best cross-validation score
print("Best Cross-validation Score: ", grid_search.best_score_)

# Test score
test_score = grid_search.score(X_test, y_test)
print("Test Score: ", test_score)
```
